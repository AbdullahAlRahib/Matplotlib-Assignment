# **Q2**

**What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.**


**Answer:**


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.show()
```

# **Q3**

Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

    import numpy as np

    For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
    For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
    For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
    For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

    
**Answer:**


```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# Plot line 1
axes[0, 0].plot(x1, y1)
axes[0, 0].set_title('Line 1')

# Plot line 2
axes[0, 1].plot(x2, y2)
axes[0, 1].set_title('Line 2')

# Plot line 3
axes[1, 0].plot(x3, y3)
axes[1, 0].set_title('Line 3')

# Plot line 4
axes[1, 1].plot(x4, y4)
axes[1, 1].set_title('Line 4')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

# **Q4**

What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

    import numpy as np
    company = np.array(["Apple", "Microsoft", "Google", "AMD"])
    profit = np.array([3000, 8000, 1000, 10000])


**Answer:**



```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
# Bar plot
plt.bar(company, profit)
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Profit by Company')
plt.show()
```

```python
# Horizontal bar plot
plt.barh(company, profit)
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.title('Profit by Company')
plt.show()
```


# **Q5**

What is a box plot? Why is it used? Using the following data plot a box plot.

    box1 = np.random.normal(100, 10, 200)
    box2 = np.random.normal(90, 20, 200)

**Answer:**


```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
fig, ax = plt.subplots()
ax.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
ax.set_xlabel('Box')
ax.set_ylabel('Value')
ax.set_title('Box Plot')
plt.show()
```
